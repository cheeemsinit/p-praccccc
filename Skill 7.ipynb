```python
# Q1 Create python program to design a simple Calculator to perform basic arithmetic operations like addition, 
#subtraction, multiplication and division.

import tkinter as tk
from functools import partial

def add(n1,n2):
    num1 = (n1.get())
    num2 = (n2.get())
    result = int(num1) + int(num2)
    label_result.config(text="Result=%d"%result)
    return

def sub(n1,n2):
    num1 = (n1.get())
    num2 = (n2.get())
    result = int(num1) - int(num2)
    label_result.config(text="Result=%d"%result)
    return

def multi(n1,n2):
    num1 = (n1.get())
    num2 = (n2.get())
    result = int(num1) * int(num2)
    label_result.config(text="Result=%d"%result)
    return


def div(n1,n2):
    num1 = (n1.get())
    num2 = (n2.get())
    result = int(num1) / int(num2)
    label_result.config(text="Result=%d"%result)
    return

root = tk.Tk()
root.geometry("400x200")
root.title("Calculator")

number1 = tk.StringVar()
number2 = tk.StringVar()

labelNum1 = tk.Label(root,text="A")
labelNum1.grid(row=1,column=0)

labelNum2 = tk.Label(root,text="B")
labelNum2.grid(row=2,column=0)

label_result = tk.Label(root)
label_result.grid(row=7,column=2)

entryNum1 = tk.Entry(root,textvariable=number1)
entryNum1.grid(row=1,column=2)

entryNum2 = tk.Entry(root,textvariable=number2)
entryNum2.grid(row=2,column=2)

addition = partial(add,number1,number2)
Subtraction = partial(sub,number1,number2)
Multiplication = partial(multi,number1,number2)
Division = partial(div,number1,number2)

buttonCal = tk.Button(root,text="Addition",command=addition)
buttonCal.grid(row=3,column=0)

buttonCal = tk.Button(root,text="Subtraction",command=Subtraction)
buttonCal.grid(row=3,column=1)

buttonCal = tk.Button(root,text="Multiplication",command=Multiplication)
buttonCal.grid(row=3,column=2)

buttonCal = tk.Button(root,text="Division",command=Division)
buttonCal.grid(row=3,column=3)

root.mainloop()
```
