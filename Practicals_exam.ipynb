```python
# Question - 2 
import numpy as np

def check_consis(coeff_matrix, const_matrix):
    """
    Function to check the consistency of a system of linear equation.
    """
    A = np.array(coeff_matrix)
    B = np.array(const_matrix).reshape(-1, 1)
    
    augmented_matrix = np.hstack((A, B))
    
    rank_A = np.linalg.matrix_rank(A)
    rank_augmented = np.linalg.matrix_rank(augmented_matrix)
    
    if rank_A == rank_augmented:
        return "The system of equations is consistent."
    else:
        return "The system of equations is inconsistent."

coefficients = [
    [1, 4, 9],  
    [2, -12, 19], 
    [7, 11, -26] 
]

constants = [-96, 45, 17]

result = check_consis(coefficients, constants)
print(result)
```

```python
# Question -4 

# plot the following functions using python 
#  sigmoid , softmax , ReLu

import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def softmax(x):
    exp_x = np.exp(x - np.max(x))  
    return exp_x / np.sum(exp_x, axis=0)

def relu(x):
    return np.maximum(0, x)



x = np.linspace(-10, 10, 500)

sigmoid_values = sigmoid(x)
softmax_values = softmax(x)  
relu_values = relu(x)

plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.plot(x, sigmoid_values, label='Sigmoid', color='blue')
plt.title("Sigmoid Function")
plt.xlabel("x")
plt.ylabel("Sigmoid(x)")
plt.grid()
plt.legend()

plt.subplot(1, 3, 2)
plt.plot(x, softmax_values, label='Softmax', color='green')
plt.title("Softmax Function")
plt.xlabel("x")
plt.ylabel("Softmax(x)")
plt.grid()
plt.legend()

plt.subplot(1, 3, 3)
plt.plot(x, relu_values, label='ReLU', color='red')
plt.title("ReLU Function")
plt.xlabel("x")
plt.ylabel("ReLU(x)")
plt.grid()
plt.legend()

plt.tight_layout()
plt.show()

```

```python
# Question - 6 
import numpy as np
from sympy import Matrix

A = np.array([
    [1, -2, 2],
    [3,  2, -3],
    [1, -3, 8]
])

x = np.array([1, 2, 3])  

y = np.dot(A, x)

print("Matrix A:")
print(A)
print("\nVector x:")
print(x)
print("\nTransformed vector y = A * x:")
print(y)

A_sympy = Matrix(A)  
kernel = A_sympy.nullspace()


print("\nKernel of the matrix A (null space):")
if kernel:
    for vec in kernel:
        print(vec)  
else:
    print("The kernel is trivial (only the zero vector).")
```
