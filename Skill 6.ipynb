```python
'''Q1 Create Python program to load the Item.csv file in DataFrame, find the missing values follows: 
(The Item.csv file is uploaded in Classroom)
Replace missing value of Quantity by median
Replace missing value of Price by Mean
Replace missing value of bought column by Standard Deviation
Replace missing value of Forenoon column with maximum value
Replace missing value of Afternoon column with minimum value
Display all data after replacement back to user.'''
```

```python
import pandas as pd
```

```python
data = pd.read_csv("Item.csv")
data
```

```python
data.info()
```

```python
data.isna()
```

```python
data.isna().sum()
```

```python
#Replace missing values of quantity columns with mean.
#Replace missing values of price columns with median.
#Replace missing values of bought columns with Standard Deviation.
#Replace missing values of forenoon columns with minimum value in that column.
#Replace missing values of afternoon columns with maximum value in that column.
```

```python
#Replace missing values of quantity columns with mean.
data['quantity'] = data['quantity'].fillna(data['quantity'].mean())
data
```

```python
#Replace missing values of price columns with median.
data['price']=data['price'].fillna(data['price'].median())
data
```

```python
#Replace missing values of bought columns with Standard Deviation.
data['bought']=data['bought'].fillna(data['bought'].std())
data
```

```python
#Replace missing values of afternoon columns with maximum value in that column.
data['afternoon']=data['afternoon'].fillna(data['afternoon'].max())
data
```
