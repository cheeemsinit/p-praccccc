```python
#Q1 Create a python program that asks the user to typeÂ a word and return him its reverse as part of label.

from tkinter import *
from functools import partial

def func1(n1):
    a1 = n1.get()
    string = ""
    string = string + a1[::-1]
    result.config(text="Result:%s"%string)
    return


a = Tk()
a.title("tk")
a.geometry("400x200")

input_string = StringVar()

word = Label(a,text="Enter the word:")
word.place(x=10,y=10)

result = Label(a)
result.place(x=160,y=40)

e1 = Entry(a,textvariable=input_string)
e1.place(x=150,y=10)

reverse = partial(func1,input_string)

b1 = Button(a,text="Validate",command=reverse)
b1.place(x=180,y=90)
#b1.place(x=10,y=100)
a.mainloop()
```

```python
# Q2 Create a python program to display 3 checkbutton with name of 3 classes like FYBSC, SYBSC,and TYBSC. Display proper
# message when user select or unselect any of the checkbutton.

from tkinter import messagebox

def fun1():
    if checkvar1.get() == 1:
        messagebox.showinfo("Hello","You selected FYBSC.")
    if checkvar1.get() == 0:
        messagebox.showinfo("Hello","You unselected FYBSC.")
        
def fun2():
    if checkvar2.get() == 1:
        messagebox.showinfo("Hello","You selected SYBSC.")
    if checkvar2.get() == 0:
        messagebox.showinfo("Hello","You unselected SYBSC.")
        
def fun3():
    if checkvar3.get() == 1:
        messagebox.showinfo("Hello","You selected TYBSC.")
    if checkvar3.get() == 0:
        messagebox.showinfo("Hello","You unselected TYBSC.")
            
```

```python
from tkinter import *

p = Tk()
p.title("Check Buttoon")
p.geometry("350x350")

checkvar1 = IntVar()

checkvar2 = IntVar()

checkvar3 = IntVar()

chkbtn1 = Checkbutton(p,text="FYBSC",variable=checkvar1,onvalue=1,offvalue=0,height=5,width=10,command=fun1)

chkbtn2 = Checkbutton(p,text="SYBSC",variable=checkvar2,onvalue=1,offvalue=0,height=5,width=10,command=fun2)

chkbtn3 = Checkbutton(p,text="TYBSC",variable=checkvar3,onvalue=1,offvalue=0,height=5,width=10,command=fun3)

chkbtn1.pack()
chkbtn2.pack()
chkbtn3.pack()

p.mainloop()
```

```python
# Q4 Create python program to demonstrate any two GUI components by adding in parent window with options like bg, 
# fg, activebackground, activeforeground, height, width, text, font, etc.

from tkinter import *
from functools import partial

def gen_sel():
    selection = "You selected option : "+str(var.get())
    result.config(text=selection)

def det(name12,lname12):
    n1 = (name12.get())
    n2 = (lname12.get())
    result2.config(text = (n1,n2)) 
    
    

from tkinter import *
top  = Tk()
top.geometry("500x500")

var = IntVar()
name12= StringVar()
lname12 = StringVar()

name = Label(top, text = "First Name")
name.grid(row=5,column=5)

l_name = Label(top,text="Last Name")
l_name.grid(row=10,column=5)

e1 = Entry(top,textvariable=name12)
e1.grid(row=5,column=10)

e2 = Entry(top,textvariable=lname12)
e2.grid(row=10,column=10)


details = partial(det,name12,lname12)


male = Radiobutton(top,text="Male",variable=var,value=1,command=gen_sel)
male.grid(row=15,column=20)

female = Radiobutton(top,text="Female",variable=var,value=2,command=gen_sel)
female.grid(row=20,column=20)



submit = Button(top,text="Submit",fg='blue',bg='pink',activebackground='white',activeforeground='black',command=details)
submit.grid(row=25,column=25)

result = Label(top)
result.grid(row=30,column=25)

result2 = Label(top)
result2.grid(row=35,column=25)

top.mainloop()
```
