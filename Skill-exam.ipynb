```python
# Skills - Questions 


import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
    return 1 / (1 + np.exp(-x))


def relu(x):
    return np.maximum(0, x)



x = np.linspace(-10, 10, 500)

sigmoid_values = sigmoid(x)
relu_values = relu(x)

plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.plot(x, sigmoid_values, label='Sigmoid', color='blue')
plt.title("Sigmoid Function")
plt.xlabel("x")
plt.ylabel("Sigmoid(x)")
plt.grid()
plt.legend()


plt.subplot(1, 3, 3)
plt.plot(x, relu_values, label='ReLU', color='red')
plt.title("ReLU Function")
plt.xlabel("x")
plt.ylabel("ReLU(x)")
plt.grid()
plt.legend()

plt.tight_layout()
plt.show()

```
